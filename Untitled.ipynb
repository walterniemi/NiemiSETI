```python
import os
import urllib.request

# Define download directory
download_dir = "gw_data"
os.makedirs(download_dir, exist_ok=True)

# Number of files to download (adjust if needed)
num_files = 5  # Change this number to download more files

# Download the first 'num_files' files
downloaded_files = []

for url in urls[:num_files]:
    filename = os.path.join(download_dir, os.path.basename(url))
    print(f"Downloading: {filename} ...")
    
    try:
        urllib.request.urlretrieve(url, filename)
        downloaded_files.append(filename)
        print(f"✅ Successfully downloaded: {filename}")
    except Exception as e:
        print(f"❌ Failed to download {filename}: {e}")

print("\nDownload complete. Files saved in:", download_dir)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
from scipy.signal import butter, filtfilt
from scipy.fftpack import fft, fftfreq

# Load the HDF5 file
file_path = "gw_data/H-H1_GWOSC_O3a_4KHZ_R1-1238163456-4096.hdf5"  # Replace with your actual file name
with h5py.File(file_path, 'r') as f:
    strain_data = f['strain']['Strain'][:]

# Remove NaN values
strain_data = np.nan_to_num(strain_data)

# Define Sampling Rate (from metadata)
sampling_rate = 4096  # LIGO default sample rate in Hz

# Butterworth Band-Pass Filter (10 Hz - 1000 Hz)
def butter_bandpass(lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a

def apply_bandpass_filter(data, lowcut=10, highcut=1000, fs=4096, order=4):
    b, a = butter_bandpass(lowcut, highcut, fs, order)
    return filtfilt(b, a, data)

# Apply the filter
filtered_strain = apply_bandpass_filter(strain_data)

# Compute the FFT
N = len(filtered_strain)
frequencies = fftfreq(N, d=1/sampling_rate)
fft_values = np.abs(fft(filtered_strain))

# Plot the Power Spectrum
plt.figure(figsize=(12, 6))
plt.plot(frequencies[:N // 2], fft_values[:N // 2])  # Only positive frequencies
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power")
plt.title("Fourier Transform of Gravitational Wave Data (Filtered)")
plt.yscale("log")  # Log scale to see small signals
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import pandas as pd
from scipy.signal import butter, filtfilt
from scipy.fftpack import fft, fftfreq

# Load the HDF5 file
file_path = "gw_data/H-H1_GWOSC_O3a_4KHZ_R1-1238163456-4096.hdf5"  # Update with the correct filename
with h5py.File(file_path, 'r') as f:
    strain_data = f['strain']['Strain'][:]

# Remove NaN values
strain_data = np.nan_to_num(strain_data)

# Define Sampling Rate (from metadata)
sampling_rate = 4096  # LIGO default sample rate in Hz

# Butterworth Band-Pass Filter (10 Hz - 1000 Hz)
def butter_bandpass(lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a

def apply_bandpass_filter(data, lowcut=10, highcut=1000, fs=4096, order=4):
    b, a = butter_bandpass(lowcut, highcut, fs, order)
    return filtfilt(b, a, data)

# Apply the filter
filtered_strain = apply_bandpass_filter(strain_data)

# Compute the FFT
N = len(filtered_strain)
frequencies = fftfreq(N, d=1/sampling_rate)
power_spectrum = np.abs(fft(filtered_strain))  # Use correct variable name

# Plot the Power Spectrum
plt.figure(figsize=(12, 6))
plt.plot(frequencies[:N // 2], power_spectrum[:N // 2])  # Only positive frequencies
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power")
plt.title("Fourier Transform of Gravitational Wave Data (Filtered)")
plt.yscale("log")  # Log scale to see small signals
plt.grid(True)
plt.show()

# **✅ Save the FFT Data to CSV**
data = np.column_stack((frequencies[:N // 2], power_spectrum[:N // 2]))  # Positive frequencies only
df = pd.DataFrame(data, columns=["Frequency (Hz)", "Power"])
csv_filename = "power_spectrum_data1.csv"
df.to_csv(csv_filename, index=False)

print(f"✅ CSV file saved successfully: {csv_filename}")
```
